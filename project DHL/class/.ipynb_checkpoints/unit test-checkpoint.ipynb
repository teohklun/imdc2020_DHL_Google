```python
def getInitRouteInfo():
    routeInfo = {}
    routeInfo["routeArray"] = []

    routeInfo["totalSeconds"] = 0
    routeInfo["totalVehicle"] = 0
    routeInfo["totalJob"] = 0
    routeInfo["jobArray"] = []
    routeInfo["timeDiffArray"] = []
    return routeInfo

def getInitDayInfo():
    dayInfo = {}
    dayInfo["routeArray"] = []
    dayInfo["totalSeconds"] = 0
    dayInfo["totalVehicle"] = 0
    dayInfo["totalJob"] = 0
    dayInfo["jobArray"] = []
    dayInfo["timeDiffArray"] = []
    return dayInfo
```

```python
# fullDataFrame = pd.read_csv("3.csv")
import datetime


morningStart = "06:00:00"
morningEnd = "13:59:59"

afternoonStart = "14:00:00"
afternoonEnd = "20:00:00"

morningSSec = datetime.datetime.strptime(morningStart, '%H:%M:%S').time()
morningESec = datetime.datetime.strptime(morningEnd, '%H:%M:%S').time()
afternoonSSec = datetime.datetime.strptime(afternoonStart, '%H:%M:%S').time()
afternoonESec = datetime.datetime.strptime(afternoonEnd, '%H:%M:%S').time()
orignalGStorage = 'gs://real-bucket-dhl/original.csv'
def getRouteInfo(fullDataFrame, dates,session = None):
#     fullDataFrame = pd.read_csv(orignalGStorage)
    if(fullDataFrame.empty):
        return "csv maybe empty"
    subDataFrame = fullDataFrame.loc[fullDataFrame["Act Dt"].isin(dates)]

    if(subDataFrame.empty):
        return "not found"
    subDataFrame['Act Tm'] = pd.to_datetime(subDataFrame['Act Tm'])
    subDataFrame['Act Tm'] = [time.time() for time in subDataFrame['Act Tm']]
    
    dayInfo = getInitDayInfo()
    for date in dates:
        singleDay = subDataFrame.loc[subDataFrame["Act Dt"] == date]

        if(session):
            morningMask = (singleDay["Act Tm"] <= morningESec) & (singleDay["Act Tm"] >= morningSSec)
            afternoonMask = (singleDay["Act Tm"] <= afternoonESec) & (singleDay["Act Tm"] >= afternoonSSec)
            
            morningFrame = singleDay.loc[morningMask]
            afternoonFrame = singleDay.loc[afternoonMask]
#             afternoonFrame = singleDay.loc[afternoonMask]
            if(session == "m"):
                singleDay = morningFrame
            elif(session =="a"):
                singleDay = afternoonFrame
            else:
                return "session not support"
        names = singleDay["Courier id"].unique()
        names = list(names)
        routeInfo = getInitRouteInfo()
        for name in names:
            route = {}
            singleRoute = singleDay.loc[singleDay["Courier id"] == name]
            if(singleRoute.empty):
                break
                
            singleRoute 
            
#             jobs = singleRoute.loc[(singleRoute["Act Ckpt Code"] != "DEPAR") & (singleRoute["Act Ckpt Code"] != "ARRVD") ]
            route["start"] = singleRoute.iloc[0]["Act Tm"]
            route["end"] = singleRoute.iloc[-1]["Act Tm"]
            jobs = (singleRoute.shape[0])-2
#             jobs = jobs.shape[0]

            dateTimeB = datetime.datetime.combine(datetime.date.today(), route["start"])
            dateTimeA = datetime.datetime.combine(datetime.date.today(), route["end"])
            dateTimeDifference = dateTimeA - dateTimeB
            dateTimeDifference = dateTimeDifference.total_seconds()
            route["diff"] = dateTimeDifference
            route["name"] = name
            route["job"] = jobs
            routeInfo["routeArray"].append(route.copy())
            routeInfo["totalSeconds"] += route["diff"]
            routeInfo["totalVehicle"] += 1
            routeInfo["totalJob"] +=jobs
            routeInfo["jobArray"].append(jobs)
            routeInfo["timeDiffArray"].append(dateTimeDifference)
        
        dayInfo["routeArray"].append(routeInfo["routeArray"])
        dayInfo["totalSeconds"] += routeInfo["totalSeconds"]
        dayInfo["totalVehicle"] += routeInfo["totalVehicle"]
        dayInfo["totalJob"] +=routeInfo["totalJob"]
        dayInfo["jobArray"].append(routeInfo["jobArray"])
        dayInfo["timeDiffArray"].append(routeInfo["timeDiffArray"])
        
    return dayInfo

fullDataFrame = pd.read_csv(orignalGStorage)
allDates = fullDataFrame["Act Dt"].unique()
allDates = list(allDates)
inputdate = [20200201, 20200204]
inputdate = allDates
a = getRouteInfo(fullDataFrame, inputdate)
a
```

```python
def getRouteInfoForOptimal(fullDataFrame, dates,session = None):
#     fullDataFrame = pd.read_csv(orignalGStorage)
    if(fullDataFrame.empty):
        return "csv maybe empty"
    subDataFrame = fullDataFrame.loc[fullDataFrame["Act Dt"].isin(dates)]

    if(subDataFrame.empty):
        return "not found"
    subDataFrame['Act Tm'] = pd.to_datetime(subDataFrame['Act Tm'])
    subDataFrame['Act Tm'] = [time.time() for time in subDataFrame['Act Tm']]
    
    dayInfo = getInitDayInfo()
    for date in dates:
        singleDay = subDataFrame.loc[subDataFrame["Act Dt"] == date]

        if(session):
            morningMask = (singleDay["Act Tm"] <= morningESec) & (singleDay["Act Tm"] >= morningSSec)
            afternoonMask = (singleDay["Act Tm"] <= afternoonESec) & (singleDay["Act Tm"] >= afternoonSSec)
            
            morningFrame = singleDay.loc[morningMask]
            afternoonFrame = singleDay.loc[~morningMask]
            afternoonFrame = singleDay.loc[afternoonMask]
            if(session == "m"):
                singleDay = morningFrame
            elif(session =="a"):
                singleDay = afternoonFrame
            else:
                return "session not support"
        batchIDs = singleDay["batchID"].unique()
        batchIDs = list(batchIDs)
        routeInfo = getInitRouteInfo()
        for batchID in batchIDs:
            route = {}
            singleRoute = singleDay.loc[singleDay["batchID"] == batchID]
            if(singleRoute.empty):
                break
                
            singleRoute 
            
            jobs = singleRoute.loc[(singleRoute["Act Ckpt Code"] != "DEPAR") & (singleRoute["Act Ckpt Code"] != "ARRVD") ]
            route["start"] = singleRoute.iloc[0]["Act Tm"]
            route["end"] = singleRoute.iloc[-1]["Act Tm"]
#             jobs = (singleRoute.shape[0])-2
            jobs = jobs.shape[0]

            dateTimeB = datetime.datetime.combine(datetime.date.today(), route["start"])
            dateTimeA = datetime.datetime.combine(datetime.date.today(), route["end"])
            dateTimeDifference = dateTimeA - dateTimeB
            dateTimeDifference = dateTimeDifference.total_seconds()
            route["diff"] = dateTimeDifference
            route["name"] = singleRoute.iloc[0]["Courier id"]
            route["job"] = jobs
            routeInfo["routeArray"].append(route.copy())
            routeInfo["totalSeconds"] += route["diff"]
            routeInfo["totalVehicle"] += 1
            routeInfo["totalJob"] +=jobs
            routeInfo["jobArray"].append(jobs)
            routeInfo["timeDiffArray"].append(dateTimeDifference)
        
        dayInfo["routeArray"].append(routeInfo["routeArray"])
        dayInfo["totalSeconds"] += routeInfo["totalSeconds"]
        dayInfo["totalVehicle"] += routeInfo["totalVehicle"]
        dayInfo["totalJob"] +=routeInfo["totalJob"]
        dayInfo["jobArray"].append(routeInfo["jobArray"])
        dayInfo["timeDiffArray"].append(routeInfo["timeDiffArray"])
        
    return dayInfo
```

```python
def convertTimeFormat(dateFrame):
    dateFrame["Act Tm"] = pd.to_datetime(dateFrame['Act Tm'])
    dateFrame["Act Tm"] = dateFrame["Act Tm"].dt.hour * 60 *60 + dateFrame["Act Tm"].dt.minute*60  + dateFrame["Act Tm"].dt.second
    dateFrame["Open"] = pd.to_datetime(dateFrame['Open'])
    dateFrame["Closed"] = pd.to_datetime(dateFrame['Closed'])
    dateFrame["Open"] = dateFrame["Open"].dt.hour * 60 *60 + dateFrame["Open"].dt.minute*60  + dateFrame["Open"].dt.second
#     dateFrame["Open"] =[ x if x == x else startSec for index, x in enumerate(dateFrame["Open"]) ]
    dateFrame["Open"] =[ x if x == x else getTimeSec(dateFrame.iloc[index]["Act Tm"], "start") for index, x in enumerate(dateFrame["Open"]) ]

    dateFrame["Closed"] = dateFrame["Closed"].dt.hour * 60 *60 + dateFrame["Closed"].dt.minute*60  + dateFrame["Closed"].dt.second
    dateFrame["Closed"] =[ x if x == x else getTimeSec(dateFrame.iloc[index]["Act Tm"], "end") for index, x in enumerate(dateFrame["Closed"]) ]
    return dateFrame

endSec = 86300
startSec = 1

def getTimeSec(second, getMorA):
    
    morningDefaultStartAcceptance = 6 * 60 *60
    morningDefaultEndAcceptance = 14 * 60 *60
    
    afternoonDefaultStartAcceptance = 14 * 60 *60
    afternoonDefaultEndAcceptance = 23 * 60 *60
    if(second > morningDefaultStartAcceptance and second < morningDefaultEndAcceptance):
        if(getMorA == "start"):
            return morningDefaultStartAcceptance
        else:
            return morningDefaultEndAcceptance

    else:
        if(getMorA == "start"):
            return afternoonDefaultStartAcceptance
        else:
            return afternoonDefaultEndAcceptance

    
    

if 'id' not in fullDataFrame .columns:
    fullDataFrame["id"] = fullDataFrame.index+1
# fullDataFrame = fullDataFrame.loc[(~(fullDataFrame['Act Ckpt Code'].isin(["CA","BA","NH"])) )  ]
dateFrame = fullDataFrame.loc[fullDataFrame["Act Dt"] == 20200203]
dateFrame = convertTimeFormat(dateFrame)

driverTspFrame = dateFrame.loc[dateFrame["Courier id"] == "rafinord"]
vehicleFrame = driverTspFrame.iloc[[0,-1]]
driverJobsFrame =driverTspFrame.iloc[1:-1]
jobsList = []
vehicleList = []
deliveryOrPick = [1]
duration = 300
subangLocation = [101.9381 ,2.7297] #should be seremban 
defaultCapacity = 99999
# for y in vehicleFrame:

customCapacity = True

vehicleList.append(Vehicle(vehicleFrame.iloc[0]["id"] + 1,subangLocation, defaultCapacity, [] , [vehicleFrame.iloc[0]["Act Tm"], vehicleFrame.iloc[1]["Act Tm"] ], subangLocation, vehicleFrame.iloc[0]["Courier id"]))

for index, y in driverJobsFrame.iterrows():
    print(y["lgtd"])
    jobsList.append(Job(y["id"],deliveryOrPick,[y["lgtd"], y["lat"]] , [], [[ int(y["Open"]) , int(y["Closed"]) ]], duration , y["Street"]  if y["Street"] == y["Street"] else "street is blank"        ))
print(jobsList[0].time_windows)
print(vehicleList[0].time_window)

dict_t = createDictionaryObjectJobsVehicles(jobsList,vehicleList)

content = json.dumps(dict_t, cls=NpEncoder)
# print(content)
# print(content)
url = 'http://35.213.166.175:3000' #vroomRoute

headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
r = requests.post(url, data=content, headers=headers)    

if (r.status_code != 200) :
    print ("Error request code : ")
    print(r.content)
    print ("error! " + str(r.status_code) + "reason : " + r.content)
else:

    def addDescriptionToStepDict(fullDF, dictStep):
        import re
        jobInformation = fullDF.loc[fullDF['id'] == int(dictStep["job"])]
        if(jobInformation["Street"].values[0] == jobInformation["Street"].values[0]):
            string = jobInformation["Street"].values[0]
            description = re.sub('\s+', ' ', string)
            dictStep["description"] = description
        else:
            dictStep["description"] = "Street is blank"
        return dictStep
    
    dict_t = json.loads(r.content)
    for route in dict_t["routes"]:
        for step in route["steps"]:
            # print(step)
            if(step["type"] == "job"):
                step = addDescriptionToStepDict(driverTspFrame, step)
            elif(step["type"] == "start"):
                name = vehicleFrame["Courier id"].iloc[0]
#                 step["name"] = name["Courier id"].values[0]
                step["name"] = name

    content = json.dumps(dict_t)
    from google.cloud import storage
    client = storage.Client()
    bucket = client.get_bucket("real-bucket-dhl")
    blob = bucket.blob("compare_driver" + ".json")

    blob.upload_from_string(content, "application/json")
#     print(dict_t["summary"]["duration"])
    startSeconds = dict_t["routes"][0]["steps"][0]["arrival"]
    endSeconds = dict_t["routes"][0]["steps"][-1]["arrival"]
    print("Optimal Start seconds : " + str(startSeconds))
    print("Optimal End Seconds : " +  str(endSeconds))
    print("Orignal Start seconds : " +  str(vehicleFrame.iloc[0]["Act Tm"]))
    print("Orignal End Seconds : " +  str(vehicleFrame.iloc[1]["Act Tm"]))
    
    
#     print(vehicleFrame["Act Tm"])
#     print(dateFrame["Open"])
```
